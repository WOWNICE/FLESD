# Convergence Analysis

```python
import matplotlib
from matplotlib import pyplot as plt

plt.style.use(['seaborn'])
plt.rcParams['figure.figsize'] = (7.5, 3)

import numpy as np
import json
import os
os.environ["KMP_DUPLICATE_LIB_OK"]="TRUE"

with open('data/our_local_losses.json', 'r') as f:
    fled_losses = json.load(f)

with open('data/fedavg_local_losses.json', 'r') as f:
    fedavg_losses = json.load(f)
    

################# OURS.
# Average and std the local losses.
avg_losses = sum([np.array(x) for x in fled_losses.values()])/len(fled_losses)
std_losses = np.std(np.array([np.array(x) for x in fled_losses.values()]), axis=1)
for i in range(5):
    ys = avg_losses[i]
    xs = range(i*400, (i+1)*400)
    plt.plot(xs, ys, '--', color='salmon', label='FLESD-local' if i == 0 else None)
    
    plt.fill_between(xs, ys-std_losses[i], ys+std_losses[i], facecolor='salmon', alpha=0.3)
#     plt.plot(xs, ys+std_losses[i], color='salmon', alpha=0.3)
#     plt.plot(xs, ys+std_losses[i], color='salmon', alpha=0.3)

for i in range(4):
    xs = [(i+1)*400-1, (i+1)*400]
    ys = [avg_losses[i][-1], avg_losses[i+1][0]]
    plt.plot(xs, ys, color='brown', label='FLESD-global' if i == 0 else None)

    
################ FEDAVG.
avg_losses = sum([np.array(x) for x in fedavg_losses.values()])/len(fedavg_losses)
tmp=np.array([np.array(x) for x in fedavg_losses.values()])
std_losses = np.std(tmp, axis=0)
print(std_losses)
print(avg_losses.shape)
print(std_losses.shape)

for i in range(20):
    ys = avg_losses[i]
    xs = range(i*100, (i+1)*100)
    plt.plot(xs, ys, '--', color='slateblue', label='FedAvg-local' if i == 0 else None, alpha=0.5, )
    
    plt.fill_between(xs, ys-std_losses[i], ys+std_losses[i], facecolor='slateblue', alpha=0.3)


for i in range(19):
    xs = [(i+1)*100-1, (i+1)*100]
    ys = [avg_losses[i][-1], avg_losses[i+1][0]]
    plt.plot(xs, ys, color='darkblue', label='FedAvg-global' if i == 0 else None)

    
plt.subplots_adjust(bottom=.2)
plt.ylabel('Loss Value')
plt.xlabel('Accumulated Local Epochs')
plt.xticks([0, 400, 800, 1200, 1600, 2000])
plt.legend(loc='best')

plt.savefig('convergence.pdf')
```
